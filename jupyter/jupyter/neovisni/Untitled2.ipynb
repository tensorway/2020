```python
import sys
from rplidar import RPLidar
import numpy as np
import time
import math 
import cv2
from robocup import simulate_lidar as sim
from robocup import field
```

```python
field_t = field.field_t
```

```python
w = 500
x0 = y0 = 1
t = 0
out = False

angles = []
for i in range (360):
    angles.append(i)

#field_t = sim.field_t
        
print(field_t)
while (x0 < w and y0 < w and not out):
    test_scan = np.zeros((w, w), np.uint8)
    for l in field_t:
        cv2.line(test_scan,(l[0], l[1]),(l[2],l[3]),255,5)
    scan, points = sim.calc_scan_for_point(x0, y0, angles,field_t)
    print(type(scan))
    for i in range (len(points)):
        point = points[i]
        if type(point) is list:
            x1 = point[0]
            y1 = point[1]
            x2 = x0
            y2 = y0
            cv2.line(test_scan,(x1,y1),(x2,y2),255,1)
        else:
            print(angles[i], "ne valja",point)
            out = True

    x0 += 1
    y0 += 1
    cv2.imshow('img', test_scan)
    if cv2.waitKey(1) != -1:
        print(cv2.waitKey(1))
        break
    time.sleep(0.01)
cv2.destroyAllWindows()
```
