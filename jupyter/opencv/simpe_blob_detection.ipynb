```python
import cv2
import numpy as np
import time

cap = cv2.VideoCapture(0)
last_print = 0
while(1):

    times = time.time()
    # Take each frame
    _, frame = cap.read()

    # Convert BGR to HSV
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # define range of blue color in HSV
    lower_blue = np.array([90,50,50])
    upper_blue = np.array([150,255,255])

    # Threshold the HSV image to get only blue colors
    mask = cv2.inRange(hsv, lower_blue, upper_blue)

    # Bitwise-AND mask and original image
    res = cv2.bitwise_and(frame,frame, mask= mask)

    cv2.imshow('frame',frame)
    cv2.imshow('mask',mask)
    cv2.imshow('res',res)
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break
    fps = 1/ (time.time()-times)
    if time.time()-last_print > 2:
        print(fps)
        last_print = time.time()
cv2.destroyAllWindows()
cap.release()
```

```python
import numpy as np 
import cv2 as cv
import time as t

cap = cv.VideoCapture(0)
cap.set(3,320);
cap.set(4,240);
low = np.array([100, 50, 50])
high = np.array([150, 255, 255])
cv.namedWindow('low')
cv.createTrackbar('h', 'low', 0, 179, no)
cv.createTrackbar('s', 'low', 0, 255, no)
cv.createTrackbar('v', 'low', 0, 255, no)
cv.namedWindow('high')
cv.createTrackbar('h', 'high', 0, 179, no)
cv.createTrackbar('s', 'high', 0, 255, no)
cv.createTrackbar('v', 'high', 0, 255, no)

lowi = np.zeros((200, 200, 3), np.uint8)
highi = np.zeros((200, 200, 3), np.uint8)
while True:
    t1 = t.time()
    _, frame = cap.read()
    h = cv.getTrackbarPos('h', 'low')
    s = cv.getTrackbarPos('s', 'low')
    v = cv.getTrackbarPos('v', 'low')
    low = np.array([h, s, v])
    lowi[:] = [h, s, v]
    lowi = cv.cvtColor(lowi, cv.COLOR_HSV2BGR)
    
    h = cv.getTrackbarPos('h', 'high')
    s = cv.getTrackbarPos('s', 'high')
    v = cv.getTrackbarPos('v', 'high')
    high = np.array([h, s, v])
    highi[:] = [h, s, v]
    highi = cv.cvtColor(highi, cv.COLOR_HSV2BGR)
    hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV)
    mask = cv.inRange(hsv, low, high)
    cv.imshow('mask', mask)
    cv.imshow('image', frame)
    cv.imshow('high', highi)
    cv.imshow('low', lowi)
    k = cv.waitKey(1)
    if k != 255:
        break
    print(1/(t.time()-t1))
    
cv.destroyAllWindows()
cap.release()
```
